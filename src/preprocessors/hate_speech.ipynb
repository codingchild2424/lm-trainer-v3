```python
sentence_path = "/workspace/Coding/lm-trainer/datasets/raw_datasets/hate_speech/sentence_df_total_40614_납품.xlsx"
sentence_error_path = "/workspace/Coding/lm-trainer/datasets/raw_datasets/hate_speech/sentence_error_df_total_29386_납품.xlsx"

import pandas as pd

sentence_df = pd.read_excel(sentence_path)
sentence_error_df = pd.read_excel(sentence_error_path)
```

```python
# concat
total_sentence_df = pd.concat([sentence_df, sentence_error_df])
total_sentence_df.head()
```

```python
total_sentence_df.shape
```

```python
# if 폭력	성적 내용	욕설	차별	위법행동	중독	부적절한 관계	자해/우울	유해-기타	의미없는 내용(키보드에서 아무 버튼이나 누른 경우) columns are all 0, then add CLEAN to label column

target_column_names = [
    '폭력',
    '성적 내용',
    '욕설',
    '차별',
    '위법행동',
    '중독',
    '부적절한 관계',
    '자해/우울',
    '유해-기타',
    '의미없는 내용(키보드에서 아무 버튼이나 누른 경우)'
]

label_list = []

for row in total_sentence_df.iterrows():

    total_sum = 0

    for target_column_name in target_column_names:
        total_sum += int(row[1][target_column_name])
    
    if total_sum == 0:
        label = 'CLEAN'
    else:
        label = 'UNCLEAN'

    label_list.append(label)

total_sentence_df['label'] = label_list
    
```

```python
# total_sentence_df label name MSG -> text
total_sentence_df = total_sentence_df.rename(columns={'MSG': 'text'})
```

```python
dst_path = "/workspace/Coding/lm-trainer/datasets/pre_datasets/hate_speech_data/hate_speech_total_70000.csv"

total_sentence_df.to_csv(dst_path, index=False)
```
