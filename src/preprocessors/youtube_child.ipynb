```python
import re

def preprocessor(text):
    # erase [] inside text
    text = re.sub(r'\[.*?\]', '', text)
    # # erase \n
    text = re.sub(r'\n', '', text)
    # # erase 원문:
    text = re.sub(r'원문:', '', text)
    # erase ( )
    text = re.sub(r'\([^)]*\)', '', text)
    # erase < >
    text = re.sub(r'<[^>]*>', '', text)

    # if 수정: appeared then erase 수정: and after
    # if '수정:' in text:
    #     text = text.split('수정:')[0]
    #     text = str(text)
    

    # # if 더 많은 이야기가 궁금하다면 구독 버튼을 눌러 친구가 되어 주세요. appeared, erase and after
    # if '더 많은 이야기가 궁금하다면 구독 버튼을 눌러 친구가 되어 주세요.' in text:
    #     text = text.split('더 많은 이야기가 궁금하다면 구독 버튼을 눌러 친구가 되어 주세요.')[0]

    # if '잠들 때 함께하는 TV 솔솔솔 그림책' in text:
    #     # erase 잠들 때 함께하는 TV 솔솔솔 그림책 only
    #     text = text.replace('잠들 때 함께하는 TV 솔솔솔 그림책', '')
    
    # under 50 characters erase
    if len(text) < 50:
        return None

    return text
```

```python
import json

pre_path = "/workspace/Coding/lm-trainer/datasets/raw_datasets/cl_data/youtube_child_pre.json"

def save_json(data, dst_path):
    with open(dst_path, 'a') as f:
        json.dump(data, f, ensure_ascii=False)
        f.write('\n')


with open(you_path, 'r') as f:
    # line by line
    for line in f:
        preprocessed = preprocessor(line)

        if preprocessed is None:
            continue
        else:
            json_data = json.loads(preprocessed)
            print(json_data['text'])

            json_format = {
                'text': str(json_data['text']) + "<|endoftext|>"
            }

            save_json(json_format, pre_path)
        
```
